```python
import pandas as pd
import os

root_path = os.getcwd()+"/drive/MyDrive/contest/dacon_hansol_llm"

train_data = pd.read_csv(os.path.join(root_path, "train.csv"))
test_data = pd.read_csv(os.path.join(root_path, "test.csv"))

submission = pd.read_csv(os.path.join(root_path, "sample_submission.csv"))
train_data.sample(3)
```

```python
import langchain
import pickle

with open(file=root_path+"/test_ans.json", mode='rb') as f:
  test_ans =pickle.load(f)

with open(file=root_path+"/test_ans_resid.json", mode='rb') as f:
  test_ans_resid =pickle.load(f)

with open(file=root_path+"/train_ans_q1.json", mode='rb') as f:
  train_ans1=pickle.load(f)

with open(file=root_path+"/train_ans_q2.json", mode='rb') as f:
  train_ans2=pickle.load(f)
```

```python
test_ans = test_ans + test_ans_resid
```

```python
q = [i['query'] for i in test_ans]
a = [i['result'] for i in test_ans]

test_data = pd.merge(test_data,
                      pd.DataFrame({"질문":q,'답변':a}),
                      on = '질문',
                      how="left")
```

```python
test_data
```

```python
test_data = test_data.iloc[test_data['id'].drop_duplicates().index,:].reset_index(drop=True)
```

```python
test_data.to_csv(root_path+'/test_ans_종합_20240216.csv', index=False)
```

```python
test_data.to_excel(root_path+'/test_ans_종합_20240216.xlsx', index=False)
```
